```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def exponential(x):
    return np.exp(x)
```

```python
y = exponential(x)

plt.plot(x, y, label='exp(x)')
plt.xlabel("Time [millisecond]")
plt.ylabel("Awesomeness")
plt.grid(True)
plt.show()
```
